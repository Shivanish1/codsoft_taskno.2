```python
import pandas as pd
import tkinter as tk 

#Function to update the input field whenever a buton is pressed
def button_click(item):
    current = input_text.get()
    input_text.set(current + str(item))

#Function to clear the input field
def button_clear():
    input_text.set("")

# Function to evaluate the expression and display the result
def button_equal():
    try:
        result = str(eval(input_text.get()))
        input_text.set(result)
    except:
        input_text.set("Error")

#Main GUI setup
root = tk.Tk()
root.title("Simple Calculator")

input_text = tk.StringVar()

#Input field where the calculations are shown
input_frame = tk.Frame(root)
input_frame.pack()

input_field = tk.Entry(input_frame, textvariable=input_text, font=('arial', 18), bd=1, insertwidth=4, width=14, borderwidth=4)
input_field.grid(row=0, column=0, columnspan=4)

#Creating the buttons
buttons_frame = tk.Frame(root)
buttons_frame.pack()

button_clear = tk.Button(buttons_frame, text='Clear', width=10, height=3, command=button_clear)
button_clear.grid(row=0, column=0)

button_divide = tk.Button(buttons_frame, text='/', width=10, height=3, command=lambda: button_click('/'))
button_divide.grid(row=0, column=1)

button_multiply = tk.Button(buttons_frame, text='*', width=10, height=3, command=lambda: button_click('*'))
button_multiply.grid(row=0, column=2)

button_subtract = tk.Button(buttons_frame, text='-', width=10, height=3, command=lambda: button_click('-'))
button_subtract.grid(row=0, column=3)

#Adding number buttons(1-9)
buttons = [
    ('7',1,0),('8',1,1),('9',1,2),
    ('4',2,0),('5',2,1),('6',2,2),
    ('1',3,0),('2',3,1),('3',3,2),
    ('0',4,1)
]

for (text, row, col) in buttons:
    button = tk.Button(buttons_frame, text= text, width= 10, height=3, command=lambda text=text: button_click(text))
    button.grid(row=row, column=col)

button_add = tk.Button(buttons_frame, text='+', width=10, height=3, command=lambda: button_click('+'))
button_add.grid(row=3, column=3)

button_equal = tk.Button(buttons_frame, text='=', width=10, height=3, command=button_equal)
button_equal.grid(row=4, column=2)

button_decimal = tk.Button(buttons_frame, text='.', width=10, height=3, command=lambda: button_click('.'))
button_decimal.grid(row=4, column=0)
    
button_zero = tk.Button(buttons_frame, text='0', width=10, height=3, command=lambda: button_click('0'))
button_zero.grid(row=4, column=1)

#Start the mainloop
root.mainloop()
```
